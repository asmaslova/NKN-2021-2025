3.	Выполните стандартизацию независимого признака и масштабирование на интервал [-1, 1] зависимого признака. Решите задачи линейной регрессии и полиномиальной регрессии для степени полинома, указанной в индивидуальном задании (4), при помощи нейронных сетей с одним нейроном и оцените качество полученных моделей по показателю, указанному в индивидуальном задании (МАЕ mean absolute error). Отследите обучение нейронных сетей, изменяя, при необходимости, гиперпараметры (функцию потерь, оптимизатор, шаг обучения и т.п.) или применяя регуляризацию.

```python
df_p = df
df_p.info()
```

```python
df_p_train = df_p[0:2449]
df_p_test = df_p[2449:4899]
x_train = df_p_train['total sulfur dioxide']
y_train = df_p_train["chlorides"]
x_test = df_p_test['total sulfur dioxide']
y_test = df_p_test["chlorides"]
```

```python
plt.scatter(x_train,y_train)
```

```python
#стандартизация
x_train1 = (x_train - np.mean(x_train))/np.std(x_train)
x_test1 = (x_test - np.mean(x_test))/np.std(x_test)
plt.scatter(x_train1, y_train)
```

```python
#масштабирование
y_train1 = -1 + (((y_train - np.min(y_train))*2)/(np.max(y_train) - np.min(y_train)))
y_test1 = -1 + (((y_test - np.min(y_test))*2)/(np.max(y_test) - np.min(y_test)))
plt.scatter(x_train1, y_train1)
np.min(y_train1), np.max(y_train1)
```

Полиномиальная регрессия

```python
model = tf.keras.Sequential( [ tf.keras.layers.Dense(1, input_shape=(1,)) ] )
```

```python
model.compile(
    loss=tf.keras.losses.mean_absolute_error,
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.25),
    metrics=['mean_absolute_error']
)
```

```python
history = model.fit(x_train1, y_train1, epochs=100)
```

```python
y_predict = model.predict(x_train1)

plt.scatter(x_train1, y_train1, label='набор данных')
plt.plot(np.sort(x_test1), y_predict[np.argsort(x_test1)], color='r', label='прогноз')
plt.legend(loc='upper left')
plt.grid()
```

7.	Стандартизуйте этот признак и визуализируйте его в соответствии с индивидуальным заданием (эмпирическая функция распределения).


```python
z = df["free sulfur dioxide"]
z = (z - np.mean(z))/np.std(z)
```

```python
def ECDF(data, x):
    counter = 0
    for v in data:
        if v <= x:
            counter += 1
    return counter / len(data)
```

```python
samples = z
npoints = 500
dz = (samples.max()-samples.min())/npoints

xlist = [samples.min()+dz*i for i in range(npoints)]
ylist = [ECDF(samples, x) for x in xlist]
```

```python
df_ECDF = pd.DataFrame(ylist, columns=['V0'],index=xlist)
df_ECDF.plot.line()
```

8.	Сформируйте набор входных данных из двух стандартизованных признаков набора данных (независимый признак и определенный признак), постройте нейронную сеть (нелинейный регресор) с количеством скрытых слоев, количеством нейронов и функцией активации, указанными в индивидуальном задании, и одним нейроном в выходном слое и обучите ее на наборе данных из двух признаков и отклика. Отследите обучение нейронной сети, изменяя, при необходимости, гиперпараметры (функцию потерь, оптимизатор, шаг обучения и т.п.) или применяя регуляризацию.

(Параметры глубокой нейронной сети: кол-во скрытых слоев – 4, кол-во нейронов в скрытом слое – 64, функция активации – гиперболический тангенс)

```python
x8, z8 = np.array((x - np.mean(x))/np.std(x)), np.array((z - np.mean(z))/np.std(z))
x8.shape, z8.shape
```

```python
feature_normalizer = tf.keras.layers.Normalization(axis=None,input_shape=(1,))
feature_normalizer.adapt(x8)
```

```python
large_model = tf.keras.Sequential([
    feature_normalizer,
    tf.keras.layers.Dense(units=64, activation='tanh'),
    tf.keras.layers.Dense(units=64, activation='tanh'),
    tf.keras.layers.Dense(units=64, activation='tanh'),
    tf.keras.layers.Dense(units=64, activation='tanh'),
    tf.keras.layers.Dense(units=1)
])

large_model.summary()
```

```python
#Скомпилируем модель, используя в качестве функции потерь  среднеквадратичную ошибку MSE с оптимизатором по умолчанию (RmsProp):
large_model.compile(loss='mse')
```

```python
history8_1 = large_model.fit(
    x8, z8,
    epochs=100,
    # уровень выводимой информации
    verbose=1,
    # проверка (валидация) на 30% обучающих данных
    validation_split = 0.3)
```

```python
plot_loss(history8_1)
```

Обучение идет, как мне кажется, не очень успешно, потому что потери снижаются недостаточно быстро, поэтому я попробую поменять функцию потерь.

```python
large_model1 = tf.keras.Sequential([
    feature_normalizer,
    tf.keras.layers.Dense(units=64, activation='tanh'),
    tf.keras.layers.Dense(units=64, activation='tanh'),
    tf.keras.layers.Dense(units=64, activation='tanh'),
    tf.keras.layers.Dense(units=64, activation='tanh'),
    tf.keras.layers.Dense(units=1)
])

large_model1.summary()
```

```python
large_model1.compile(loss='mae')
```

```python
history8_2 = large_model.fit(
    x8, z8,
    epochs=100,
    # уровень выводимой информации
    verbose=1,
    # проверка (валидация) на 30% обучающих данных
    validation_split = 0.3)
```

```python
plot_loss(history8_2)
```
