## 2.1
–ù–∞—Ä–∏—Å—É–π—Ç–µ –¥–≤–∞ –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞. –ù–∞–π–¥–∏—Ç–µ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏, –¥–ª—è —á–µ–≥–æ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é arctan2. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏–π –Ω–∞ –ø—Ä–∏–º–µ—Ä–∞—Ö –æ—Å—Ç—Ä–æ–≥–æ, —Ç—É–ø–æ–≥–æ –∏ –ø—Ä—è–º–æ–≥–æ —É–≥–ª–æ–≤.


```python
x1 = 1
y1 = 1
x2 = 2
y2 = -3
n = 10

vector1 = np.array([x1, y1])
vector2 = np.array([x2, y2])
```

```python
plt.figure(figsize=(8, 8), dpi=200)  

plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='–í–µ–∫—Ç–æ—Ä 1')

plt.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='red', label='–í–µ–∫—Ç–æ—Ä 2')

angle_radians = np.arctan2(np.linalg.det([vector1, vector2]), np.dot(vector1, vector2))

angle_degrees = np.degrees(angle_radians) 

plt.text(0.5, 3.5, f'–£–≥–æ–ª –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏: {angle_degrees:.2f} –≥—Ä–∞–¥—É—Å–æ–≤', fontsize=11)

plt.xlim(-n, n)  
plt.ylim(-n, n)  
plt.axhline(0, color='black',linewidth=0.5)  
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)  
plt.legend()  
plt.title('–í–µ–∫—Ç–æ—Ä—ã –∏ —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏')  
plt.xlabel('X') 
plt.ylabel('Y')  
plt.show()
```

```python
transformation_matrix = np.array([[1, 0], [0, -1]])

trans_vector1 = transformation_matrix @ vector1
trans_vector2 = transformation_matrix @ vector2
```

```python
plt.figure(figsize=(8, 8), dpi=200)  

plt.quiver(0, 0, trans_vector1[0], trans_vector1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='–í–µ–∫—Ç–æ—Ä 1')

plt.quiver(0, 0, trans_vector2[0], trans_vector2[1], angles='xy', scale_units='xy', scale=1, color='red', label='–í–µ–∫—Ç–æ—Ä 2')

trans_angle_radians = np.arctan2(np.linalg.det([trans_vector1, trans_vector2]), np.dot(trans_vector1, trans_vector2))

trans_angle_degrees = np.degrees(trans_angle_radians) 

plt.text(0.5, 3.5, f'–£–≥–æ–ª –º–µ–∂–¥—É –≤–µ–∫—Ç–æ—Ä–∞–º–∏: {trans_angle_degrees:.2f} –≥—Ä–∞–¥—É—Å–æ–≤', fontsize=11)

plt.xlim(-n, n)  
plt.ylim(-n, n)  
plt.axhline(0, color='black',linewidth=0.5)  
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)  
plt.legend()  
plt.title('–í–µ–∫—Ç–æ—Ä—ã –∏ —É–≥–æ–ª –º–µ–∂–¥—É –Ω–∏–º–∏')  
plt.xlabel('X') 
plt.ylabel('Y')  
plt.show()
```

–ù–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è —Ä–∏—Å—É–µ—Ç –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫ –ø–æ –∑–∞–¥–∞–Ω–Ω—ã–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º –≤—Å–µ—Ö –µ–≥–æ –≤–µ—Ä—à–∏–Ω. –î–æ–ª–∂–Ω–∞ –±—ã—Ç—å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –∑–∞–¥–∞–≤–∞—Ç—å –ª—é–±–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä—à–∏–Ω –≤ –≤–∏–¥–µ –º–∞—Å—Å–∏–≤–∞ —Ç–æ—á–µ–∫.

```python
vertices = np.array([(0,0), (1, -3), (4, -2), (4, 1), (2, 2)])

def plot_the_polygon(vertices):
    fig, ax = plt.subplots(figsize=(5, 5), dpi=200)
    polygon = Polygon(vertices, closed=True, fill=False, edgecolor='b', linewidth=2)
    ax.add_patch(polygon)

    x_coords = vertices[:, 0]
    y_coords = vertices[:, 1]
    ax.plot(x_coords, y_coords, 'ro', markersize=7) # –∑–¥–µ—Å—å –º–µ–Ω—è—Ç—å –º–∞—Ä–∫–µ—Ä—ã –¥–ª—è —Ç–æ—á–µ–∫

    plt.xlim(-5, 5)  
    plt.ylim(-5, 5)  
    plt.axhline(0, color='black', linewidth=0.5)  
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)  
    plt.title('–ü—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

plot_the_polygon(vertices)
```

## 3.1

–í—ã—á–∏—Å–ª–∏—Ç–µ –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é –ø–ª–æ—â–∞–¥—å –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å –≤—ã—á–∏—Å–ª–µ–Ω–∏–π –Ω–∞ –ø—Ä–∏–º–µ—Ä–µ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞, –ø–ª–æ—â–∞–¥—å –∫–æ—Ç–æ—Ä–æ–≥–æ –º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏. –ü—Ä–æ—Å—Ç–µ–π—à–∏–π –ø—Ä–∏–º–µ—Ä –∏–∑–æ–±—Ä–∞–∂–µ–Ω –Ω–∞ —Ä–∏—Å—É–Ω–∫–µ 5.

```python
def oriented_area(vertices):
    n = len(vertices)
    area = 0
    for i in range(n):
        j = (i + 1) % n
        area += vertices[i][0] * vertices[j][1]
        area -= vertices[j][0] * vertices[i][1]
    return 0.5 * area

area = oriented_area(vertices)
print("–û—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –ø–ª–æ—â–∞–¥—å –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ =", area)
```

–ü–æ–∫–∞–∂–∏—Ç–µ —Å—Ç—Ä–µ–ª–∫–∞–º–∏ –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –æ–±—Ö–æ–¥–∞ –≤–µ—Ä—à–∏–Ω, –∫–∞–∫ —ç—Ç–æ –ø–æ–∫–∞–∑–∞–Ω–æ –Ω–∞ —Ä–∏—Å—É–Ω–∫–∞—Ö 5 –∏ 6.

```python
def plot_the_polygon_oriented(vertices):
    x_coords = vertices[:, 0]
    y_coords = vertices[:, 1]

    fig, ax = plt.subplots(figsize=(5, 5), dpi=200)

    for i in range(len(x_coords)):
        if (i==len(x_coords)-1):
            ax.quiver(x_coords[i], y_coords[i], x_coords[0]-x_coords[i], y_coords[0]-y_coords[i], angles='xy', scale_units='xy', scale=1, color='blue')
        else: 
            ax.quiver(x_coords[i], y_coords[i], x_coords[i+1]-x_coords[i], y_coords[i+1]-y_coords[i], angles='xy', scale_units='xy', scale=1, color='blue')

    plt.xlim(-5, 5)  
    plt.ylim(-5, 5)  
    plt.axhline(0, color='black', linewidth=0.5)  
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)  
    plt.title('–ü—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã–π –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

plot_the_polygon_oriented(vertices)
```

# –ó–∞–¥–∞–Ω–∏–µ 4

–ù–∞—Ä–∏—Å—É–π—Ç–µ –≥—Ä–∞—Ñ–∏–∫ —Ü–∏–∫–ª–æ–∏–¥—ã –¥–ª—è –¥–≤—É—Ö –ø–æ–ª–Ω—ã—Ö –æ–±–æ—Ä–æ—Ç–æ–≤ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏. –û—Ç–æ–±—Ä–∞–∑–∏—Ç–µ –µ–¥–∏–Ω–∏—á–Ω—ã–µ –∫–∞—Å–∞—Ç–µ–ª—å–Ω—ã–µ –∏ –Ω–æ—Ä–º–∞–ª—å–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã –≤ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö —Ç–æ—á–∫–∞—Ö —Ü–∏–∫–ª–æ–∏–¥—ã.

```python
PI = 3.1415926535
```

```python
r = 2
t = np.linspace(0, 4 * np.pi, 100)

matrix = np.array([0, -1, 1, 0]).reshape(2,2)

x = r * (t - np.sin(t))
y = r * (1 - np.cos(t))

dx = r - r * np.cos(t)
dy = r * np.sin(t)

nx = []*100
ny = []*100

for i in range(0, 100):
    nx.append(np.array(np.dot(matrix, np.array([dx[i], dy[i]])))[0])
    ny.append(np.array(np.dot(matrix, np.array([dx[i], dy[i]])))[1])

fig, ax = plt.subplots(figsize=(5, 3), dpi=200)

ax.plot(x, y, linewidth=1, color='black', label=r'–¶–∏–∫–ª–æ–∏–¥–∞')
ax.quiver(x[::6], y[::6], dx[::6], dy[::6], units='xy', angles='xy', color='blue', width=0.1, scale=2, label='–ì–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –∫–∞—Å–∞—Ç–µ–ª—å–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã')
ax.quiver(x[::6], y[::6], nx[::6], ny[::6], units='xy', angles='xy', color='red', width=0.1, scale=2, label='–í–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–µ –∫–∞—Å–∞—Ç–µ–ª—å–Ω—ã–µ –≤–µ–∫—Ç–æ—Ä—ã')


plt.xlim(0, 15)  
plt.ylim(0, 9)  
plt.axhline(0, color='black',linewidth=0.5)  
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)  
plt.title('–¶–∏–∫–ª–æ–∏–¥–∞')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

- –ö–∞—Ä—Ç–∏–Ω–∫–∞ –¥–æ–ª–∂–Ω–∞ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –ø–µ—Ä–µ—Ä–∏—Å–æ–≤—ã–≤–∞—Ç—å—Å—è –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ ùëõ

```python
n = 5
R = 2.0  
r = R * np.cos(np.pi / n)  

theta = np.linspace(0, 2 * np.pi, n, endpoint=False)
x_coords = R * np.cos(theta)
y_coords = R * np.sin(theta)

fig = plt.figure(figsize=(8, 8), dpi=200)

circle_outside = plt.Circle((0, 0), R, color='blue', fill=False, linestyle='dashed', linewidth=2)
plt.gca().add_patch(circle_outside)

circle_inside = plt.Circle((0, 0), r, color='red', fill=False, linestyle='dashed', linewidth=2)
plt.gca().add_patch(circle_inside)

plt.plot(x_coords, y_coords, color='black', marker='o', markersize=6)

for i, (x, y) in enumerate(zip(x_coords, y_coords), start=1):
    plt.text(x, y, f'$P_{{{i}}}$', fontsize=12, ha='right', va='bottom')

for i in range(n):
    plt.plot([x_coords[i], x_coords[(i + 1) % n]], [y_coords[i], y_coords[(i + 1) % n]], color='black', linewidth=3)

plt.xlim(-R - 1, R + 1)  
plt.ylim(-R - 1, R + 1)  
plt.gca().set_aspect('equal', adjustable='box')  
plt.grid(True)

plt.title(f'–ü—Ä–∞–≤–∏–ª—å–Ω—ã–π {n}-—É–≥–æ–ª—å–Ω–∏–∫ –∏ –µ–≥–æ –æ–∫—Ä—É–∂–Ω–æ—Å—Ç–∏') 
plt.xlabel('X')  
plt.ylabel('Y')  
plt.show()
```
